```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_path = 'Query_2_result.csv'

# Load the CSV file
df = pd.read_csv(file_path)

# Display the first few rows of the dataset
print(df.head())
```

```python

# Visualization 1: Skill Frequency
skill_counts = df['skills'].value_counts()
plt.figure(figsize=(12, 6))
skill_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Frequency of Skills Across Job Listings', fontsize=14)
plt.xlabel('Skills', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Visualization 2: Salary Distribution by Company
company_salary = df.groupby('company_name')['salary_year_avg'].mean().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
company_salary.plot(kind='bar', color='orange', edgecolor='black')
plt.title('Average Salary by Company', fontsize=14)
plt.xlabel('Company Name', fontsize=12)
plt.ylabel('Average Salary (Yearly)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Visualization 3: Job Title Count
job_title_counts = df['job_title'].value_counts()
plt.figure(figsize=(10, 6))
job_title_counts.plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title('Job Title Frequency', fontsize=14)
plt.xlabel('Job Titles', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
