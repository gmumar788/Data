```python
file_path = 'Query_5.csv'
# Load the CSV file
df = pd.read_csv(file_path)

# Display the first few rows of the dataset
print(df.head())
```

```python
# Visualization: Top 10 Skills by Demand Count
top_demand = df.nlargest(10, 'demand_count')
plt.figure(figsize=(12, 6))
plt.bar(top_demand['skills'], top_demand['demand_count'], color='skyblue', edgecolor='black')
plt.title('Top 10 Skills by Demand Count', fontsize=14)
plt.xlabel('Skills', fontsize=12)
plt.ylabel('Demand Count', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Visualization: Top 10 Skills by Average Salary
top_salary = df.nlargest(10, 'avg_salary')
plt.figure(figsize=(12, 6))
plt.bar(top_salary['skills'], top_salary['avg_salary'], color='lightgreen', edgecolor='black')
plt.title('Top 10 Skills by Average Salary', fontsize=14)
plt.xlabel('Skills', fontsize=12)
plt.ylabel('Average Salary (USD)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
