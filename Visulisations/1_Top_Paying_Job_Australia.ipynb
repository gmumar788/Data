```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_path = 'Query_1_result.csv'

# Load the CSV file
df = pd.read_csv(file_path)

# Display the first few rows of the dataset
print(df.head())
```

```python

# Visualization 1: Salary Distribution
plt.figure(figsize=(10, 6))
plt.hist(df['salary_year_avg'], bins=10, edgecolor='black')
plt.title('Salary Distribution', fontsize=14)
plt.xlabel('Salary (Yearly Average)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Visualization 2: Job Salaries by Posting Date
plt.figure(figsize=(12, 6))
plt.scatter(df['job_posted_date'], df['salary_year_avg'], color='blue', alpha=0.7)
plt.title('Job Salaries by Posting Date', fontsize=14)
plt.xlabel('Job Posted Date', fontsize=12)
plt.ylabel('Salary (Yearly Average)', fontsize=12)
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.show()

# Visualization 3: Top 5 Salaries by Company
top_salaries = df.nlargest(10, 'salary_year_avg')
plt.figure(figsize=(10, 6))
plt.bar(top_salaries['company_name'], top_salaries['salary_year_avg'], color='orange', edgecolor='black')
plt.title('Top 10 Salaries by Company', fontsize=14)
plt.xlabel('Company', fontsize=12)
plt.ylabel('Salary (Yearly Average)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
