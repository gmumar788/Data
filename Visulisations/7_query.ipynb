```python
import pandas as pd
import matplotlib.pyplot as plt

# Correcting the data and ensuring equal lengths for job locations and counts


file_path = 'Query_7_result.csv'

# Load the CSV file
df = pd.read_csv(file_path)


# Sort by job_count and select the top 10
df_top10 = df.sort_values(by="job_count", ascending=False).head(10)

# Plot the data
plt.figure(figsize=(12, 7))
plt.bar(df_top10["company_name"], df_top10["job_count"], alpha=0.7, color='blue')
plt.title("Top 10 Companies Hiring in Australia", fontsize=16)
plt.xlabel("Company Name", fontsize=12)
plt.ylabel("Job Count", fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.tight_layout()

# Show the plot
plt.show()
```
