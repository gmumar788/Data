```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'Query_3_result.csv'

# Load the CSV file
df = pd.read_csv(file_path)

# Display the first few rows of the dataset
print(df.head())
```

```python
# Visualization: Demand Count of Skills
plt.figure(figsize=(10, 6))
plt.bar(df['skills'], df['demand_count'], color='lightblue', edgecolor='black')
plt.title('Demand Count of Skills', fontsize=14)
plt.xlabel('Skills', fontsize=12)
plt.ylabel('Demand Count', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
