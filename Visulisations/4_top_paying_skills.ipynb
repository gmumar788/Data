```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'Query_4.csv'
# Load the CSV file
df = pd.read_csv(file_path)

# Display the first few rows of the dataset
print(df.head())
```

```python

# Visualization: Top 10 Skills by Average Salary
top_skills = df.nlargest(10, 'avg_salary')
plt.figure(figsize=(12, 6))
plt.bar(top_skills['skills'], top_skills['avg_salary'], color='skyblue', edgecolor='black')
plt.title('Top 10 Skills by Average Salary', fontsize=14)
plt.xlabel('Skills', fontsize=12)
plt.ylabel('Average Salary (USD)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
