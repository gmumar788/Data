```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'Query_8_result.csv'

# Load the CSV file
df = pd.read_csv(file_path)


# Plotting the data
plt.figure(figsize=(10, 6))
plt.barh(df["company_name"], df["job_count"], color='skyblue')
plt.xlabel("Job Count")
plt.ylabel("Company Name")
plt.title("Top Companies Hiring Data Analysts in Australia")
plt.gca().invert_yaxis()  # To display the highest value on top
plt.tight_layout()
plt.show()
```
